```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/cleaned.csv')
```

```python
df
```

```python
df['year'].replace({'TV Movie 2019': 2019}, inplace = True)
```

```python
df['year'] = df['year'].astype('int')
```

```python
df.info()
```

```python
df_year = df.groupby('year').count()
```

```python
df_year
```

```python
plt.subplots(figsize = (168, 40))
plt.title('Number of Movies Released Over Time', fontsize=28)
sns.barplot(x = df_year.index, y = df_year['title'])
```

```python
df_females_allages_avg_vote = df.groupby('females_allages_avg_vote').count()
```

```python
df_females_allages_avg_vote
```

```python
plt.subplots(figsize = (20, 8))
plt.title('Movie Counts per Female Average Vote', fontsize=28)
sns.barplot(x = df_females_allages_avg_vote.index, y = df_females_allages_avg_vote['title'])
```

```python
df_males_allages_avg_vote = df.groupby('males_allages_avg_vote').count()
```

```python
plt.subplots(figsize = (20, 8))
plt.title('Movie Counts per Male Average Vote', fontsize=28)
sns.barplot(x = df_males_allages_avg_vote.index, y = df_males_allages_avg_vote['title'])
```

```python
df_genre = df[df['genre'].notna()]
```

```python
df_genre
```

```python
df_romance = df_genre[df_genre['genre'].str.contains('Romance')]
```

```python
df_romance_group = df_romance.groupby('year').count()
```

```python
df_romance_group
```

```python
plt.subplots(figsize = (20, 8))
plt.title('Romance Movies Count Over Time', fontsize=28)
sns.barplot(x = df_romance_group.index, y = df_romance_group['genre'])
```
